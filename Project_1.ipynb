```python
import pandas as pd
df = pd.read_csv('Weather_Data.csv')
```

# 1. Find all unique 'wind speed' values in data

```python
df.head()
var = df['Wind Speed_km/h'].nunique()
print('Ans:', var)
```

# 2. Find the number of times when the Weather is exactly clear

```python
df.head()
var=df[df['Weather']=="Clear"]
var.value_counts()
print('ans:', var)

# or can use groupby
df.groupby('Weather').get_group('Clear')
```

# 3. Find the number of times when the 'wind speed is exactly 4km/h'

```python
df.head()
df[df['Wind Speed_km/h'] == 4]

#or

df.groupby('Wind Speed_km/h').get_group(4)
```

# 4. Find the all null values from the dataset

```python
df.head()
df[df.isnull()]
# df[df.notnull()] # or use notnull
var = len(df)
print("Answer of Null Values:",var)
```

# 9. Find all instances when Snow was recorded?

```python
df.head()
snow=df[df['Weather Condition']=='Snow']
var = len(snow)
print("Total no. of columns that have snow listed:",var)
```

# 10. Find all instances when Wind Speed is above 24 and Visibility is 25

```python
df.head()
var=df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km']==25)]
len(var)
```

# 11. What is the mean value of each column against each weather condition?

```python
df.head()
df.groupby('Weather Condition').mean()
```

# 12. What is the minimum and maximum value of each column against each weather condition?

```python
df.head()
min=df.groupby("Weather Condition").min()
print(min)

max=df.groupby("Weather Condition").max()
print(max)
```

# 14. Find all instances when 'Weather is clear' or 'Visibility is above 40'

```python
df.head()
var=df[(df['Weather Condition']=='Clear') | (df['Visibility_km']>40)]
len(var)
```
